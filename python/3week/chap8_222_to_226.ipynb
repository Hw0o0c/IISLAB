```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
```

```python
plt.title('Line')
```

```python
plt.plot([1,2,3],[1,2,3], marker='o')
plt.show()
```

```python
plt.title('Legend')
plt.plot([1,2,3,4], label='asc')
plt.plot([4,3,2,1], label='desc')
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3],[1,2,3], color='red')
plt.show()
```

```python
plt.plot([1,2,3],[1,2,3])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.plot([1,2,3,4],color='r', linestyle='--', label='dashed')
plt.plot([4,3,2,1],color='g', linestyle=':', label='dotted')
plt.legend()
plt.show()
```
