```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np

file_path="C:/data/titanic.csv"
df = pd.read_csv(file_path)
```

```python
df
```

```python
df_age = pd.qcut(df["Age"], q=3, labels = ["boy", "man", "old man"])
```

```python
df_age
```

```python
df["Age_class"]=df_age
```

```python
df
```

```python
df1 = df.groupby(['Sex','Age_class'])["Survived"].mean()
```

```python
df1
```

```python
df2 = df.pivot_table(['Survived'], index=['Sex','Age_class'], aggfunc='mean')
```

```python
df2.plot()
```

```python
df.plot()
```

```python
sns.catplot(x='Survived', y='Sex', hue='Age_class', kind'bar', data=df2)
```
