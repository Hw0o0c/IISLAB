```python
def bar_chart(number):
    sum =df.iloc[number, 3:9]
    df_a=pd.DataFrame([sum])
    df_a = df_a.transpose()
    df_a.index = df.iloc[0,3:9]
    df_a.columns = ["Sum"]
    df_a = df_a.apply(pd.to_numeric)
    sns.barplot(x="Sum", y=df_a.index, data=df_a)
```

```python
bar_chart(3)
```
